# Plotting in Python

## hvPlot

Adapter from [hvPlot guide](https://hvplot.pyviz.org/getting_started/index.html)

Create a dataframe

```python
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
index = pd.date_range('1/1/2000', periods=1000)
df = pd.DataFrame(np.random.randn(1000, 4), index=index,
                  columns=list('ABCD')).cumsum()
```

Plot using matplotlib

```python
df.plot()
```

Plot using bokeh

```python
import hvplot.pandas

df.hvplot()
```
